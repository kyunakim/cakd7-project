```python
user_dict = {}
def makeaccount():
    while True:
        log_id = input('Enter ID  : ')
        if log_id == -1:
            break
        log_pw = input('Enter Password : ')

        if log_id in user_dict:
            print('already exists')

        else:
            user_dict[log_id]=[log_pw,0]
            return
```
